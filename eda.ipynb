```
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```
file=pd.read_csv('data-set/preprocessed.csv')
```

```
sns.histplot(x=file['SalePrice'])
```

```
sns.kdeplot(x=file['SalePrice'])
```

```
sns.scatterplot(x=file['LotArea'], y=file['SalePrice'])
```

```
sns.scatterplot(x=file['MSSubClass'], y=file['SalePrice'])
```

```
sns.heatmap(file.corr(numeric_only=True), cmap='coolwarm')
```

```
numeric_columns=file.select_dtypes(include='number').columns
numeric_columns=numeric_columns.drop(labels=['Id'])
print(f"Analyzing the correlation of Sale Price with different features:")
for column in numeric_columns:
    print(f"{column}:{file['SalePrice'].corr(file[column])}")
```
