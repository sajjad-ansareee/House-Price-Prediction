```python
cols_to_check = [col for col in numeric_columns if col not in ['Id', 'SalePrice']]

# nan_policy would ignore the current null values present in the data set
z_scores = np.abs(stats.zscore(file[cols_to_check], nan_policy='omit'))

# there are many luxury houses that are considered to be outlier with threshold 3 so taken threshold 5
threshold = 5

# indices of rows that have at least one outlier
outlier_mask = (z_scores > threshold).any(axis=1)
outlier_rows = file[outlier_mask]

print(f"Total rows with at least one outlier: {len(outlier_rows)}")
```

```python
file=file[~outlier_mask]
```
