```python
# number of rows with atleast one null value
file.isnull().any(axis=1).sum()
```

```python
# number of columns with atleast one null value
file.isnull().any(axis=0).sum()
```

```python
# columns with atleast one null value
null_columns=file.columns[file.isnull().any(axis=0)]
```

```python
# without the .columns at end it would return a data frame
numeric_columns=file.select_dtypes(include=['number']).columns
# get the numeric columns with null values
numeric_null_columns=null_columns.intersection(numeric_columns)
```

```python
categorical_columns=file.select_dtypes(include=['str', 'object']).columns
categorical_null_columns=null_columns.intersection(categorical_columns)
```

```python
# filling the numeric null columns with there mean value
file[numeric_null_columns]=file[numeric_null_columns].fillna(file[numeric_columns].mean())
```

```python
# now the numeric columns are not having any null value
file.isnull().any(axis=0).sum()
```

```python
# filling the categorical null columns with the mode value
file[categorical_null_columns]=file[categorical_null_columns].fillna(file[categorical_columns].mode().iloc[0])
```

```python
# now there are no null values in the data set
file.isnull().any(axis=0).sum()
```

```python
# store the preprocessed csv file
file.to_csv('data-set/preprocessed.csv', index=False)
```
